```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cal_df  = pd.read_excel(r"C:\Users\rosem\Downloads\Paper_works\Model Comparison Results.xlsx", sheet_name= 'Comparison_Cal')
cal_df
```

```python
tex_df  = pd.read_excel(r"C:\Users\rosem\Downloads\Paper_works\Model Comparison Results.xlsx", sheet_name= 'Comparison_Tex')
tex_df 
```

```python
# Filter 80-20 rows
cal_80 = cal_df[cal_df['Train-Test Split'] == '80-20']
tex_80 = tex_df[tex_df['Train-Test Split'] == '80-20']

# Get MAPE columns for Demand & CO2
cal_mape_demand = cal_80.set_index('Model')['MAPE_Demand(%)']
cal_mape_co2 = cal_80.set_index('Model')['MAPE_CO2(%)']
tex_mape_demand = tex_80.set_index('Model')['MAPE_Demand(%)']
tex_mape_co2 = tex_80.set_index('Model')['MAPE_CO2(%)']

# Function to plot
def plot_mape(mape_series, title):
    plt.figure(figsize=(8, 4), dpi=80)
    bars = plt.bar(mape_series.index, mape_series.values,
                   color=['#4FC3F7', '#FF9800', '#AB47BC', '#81C784', '#D72631'])
    plt.title(title, fontsize=14)
    plt.ylabel('MAPE (%)', fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, height + 0.03,
                 f'{height:.2f}', ha='center', va='bottom', fontsize=10)
    plt.tick_params(axis='both', labelsize=14)
    plt.xticks(rotation=10)
    plt.tight_layout()
    plt.show()

#  Plots
plot_mape(cal_mape_demand, 'California  Demand MAPE (%) Comparison (80-20 Split)')
plot_mape(cal_mape_co2, 'California  CO MAPE (%) Comparison (80-20 Split)')
plot_mape(tex_mape_demand, 'Texas  Demand MAPE (%) Comparison (80-20 Split)')
plot_mape(tex_mape_co2, 'Texas  CO MAPE (%) Comparison (80-20 Split)')
```
